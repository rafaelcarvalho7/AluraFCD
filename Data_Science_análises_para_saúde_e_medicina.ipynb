```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
  return data.groupby("Country/Region").sum().iloc[:,-1]

latest_by_country(cases_confirmed).head()
```

```python
informations = [latest_by_country(cases_confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
```

```python
# Taxa de Letalidade

sum_up = combined.sum()
letality_rate1= sum_up['deaths']/sum_up['confirmed']
print(letality_rate1*100)
```

```python
letality = combined['deaths']/combined['confirmed'] *100
combined['letality_rate1'] = letality
combined
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
letality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

```python
combined['letality_rate2'] = letality_rate_2*100
combined.head()
```

```python
import seaborn as sns

sns.scatterplot(data = combined, x="confirmed", y="letality_rate1")
```

```python
import seaborn as sns

sns.scatterplot(data = combined.query("confirmed > 40"), x="confirmed", y="letality_rate1")
```

```python
sns.scatterplot(data = combined.query("confirmed > 40 and confirmed < 50000"), x="confirmed", y="letality_rate1")
sns.scatterplot(data = combined.query("confirmed > 40 and confirmed < 50000" ), x="confirmed", y="letality_rate2")
```
